```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv("data.csv")
df
```

```python
df['experience'].fillna("zero")
```

```python
df['experience']=df['experience'].fillna("zero")
df
```

```python
from word2number import w2n
df['experience'] = df['experience'].apply(w2n.word_to_num)
```

```python
df
```

```python
import math
df['test_score(out of 10)']= df['test_score(out of 10)'].fillna(math.floor(df['test_score(out of 10)'].median()))
df
                                                                
```

```python
plt.scatter(df["experience"],df["salary($)"])
plt.scatter(df["test_score(out of 10)"],df["salary($)"],color="red",marker="*")
plt.scatter(df["interview_score(out of 10)"],df["salary($)"],color="green", marker="+")
plt.xlabel("Factors")
plt.ylabel("Salary")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["experience","test_score(out of 10)","interview_score(out of 10)"]], df['salary($)'])
```

```python
reg.predict([[2,9,6]])
```
